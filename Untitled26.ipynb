```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import BertTokenizer, BertModel
import numpy as np
```

```python

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)

def get_attention_weights(text):
    tokens = tokenizer(text, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = bert_model(**tokens)
    attention = outputs.attentions[-1].squeeze(0).mean(dim=0)  # Last layer, mean over heads
    return tokens, attention

def visualize_attention(text):
    tokens, attention = get_attention_weights(text)
    token_list = tokenizer.convert_ids_to_tokens(tokens['input_ids'].squeeze().tolist())

    plt.figure(figsize=(10, 5))
    sns.heatmap(attention.numpy(), xticklabels=token_list, yticklabels=token_list, cmap='coolwarm')
    plt.title(f'Attention Heatmap for: "{text}"')
    plt.show()

```

```python
# Example 1: Medical report with severe symptoms
text1 = "Patient reports persistent cough, high fever, and difficulty breathing for the past three days."
visualize_attention(text1)

# Example 2: Milder symptoms
text2 = "Mild headache and occasional dizziness, but no fever or cough."
visualize_attention(text2)

# Example 3: Introducing rare but important symptoms
text3 = "Patient experiences sudden chest pain and loss of consciousness."
visualize_attention(text3)
```
